Now, we have to find the total cases in India

```python
total = hotel_new['Total Cases'].sum()
```

Now, we have to find the total active cases in India

```python
active = hotel_new['Active'].sum()
```

Now, we have to find the total death cases in India

```python
death = hotel_new['Deaths'].sum()
```

```python
death
```

Now, we have to find the total discharge cases in India

```python
discharge = hotel_new['Discharged'].sum()
```

```python
discharge
```

Now, we have to show the data with some graphs. Firstly we use Pie charts to show the relation between the factors 

```python
values = [total, active,discharge, death]
```

```python
lab = ['Total cases', 'Active cases','Discharge','Death Cases' ]
```

```python
plt.pie(values, labels = lab, autopct='%0.0f%%', shadow=True, explode = [0,0.1,0.1,0]);
```

And Now, we use barplot to show the relation between the factors

```python
sns.barplot(x=lab, y=values);
```
